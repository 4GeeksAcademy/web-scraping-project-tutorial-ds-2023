```python
import os
from sqlalchemy import create_engine
import pandas as pd
from dotenv import load_dotenv
import requests
from sqlalchemy import Table, Column, Integer, String, MetaData, ForeignKey
from sqlalchemy import inspect
from bs4 import BeautifulSoup 

# load the .env file variables
load_dotenv()







# Make a GET request to the website
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")



# Find all the <a> tags
tables = soup.find_all("table")



table_data = []  # List to store dictionaries representing rows

for table in tables:
    # Find all <th> tags to use as dictionary keys
    headers = [header.text.strip() for header in table.find_all("th")]

    # Find all <tr> tags (rows) in the table
    rows = table.find_all("tr")
    
    # Loop through rows and transform data into dictionaries
    for row in rows:
        # Find all <td> tags (cells) in the row
        cells = row.find_all("td")
        
        # Extract text content from cells and create a dictionary
        row_data = {}
        for i in range(min(len(headers), len(cells))):
            row_data[headers[i]] = cells[i].text.strip()
        
        # Add the dictionary to the table data list
        table_data.append(row_data)

# Print the transformed data as dictionaries
for row_dict in table_data:
    print(row_dict)








# for table in tables:
#     # Your code to process each table goes here
#     # For example, you can extract rows and cells within each table
#     rows = table.find_all("tr")  # Find all rows in the table
#     for row in rows:
#         cells = row.find_all("td")  # Find all cells in the row
#         for cell in cells:
#             print(cell.text)  # Print the





# # Transform links into a list of dictionaries
# links_data = []
# for link in links:
#     link_data = {
#         "text": link.text,
#         "url": link.get("href")
#     }
#     links_data.append(link_data)







resource_url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

response = requests.get(resource_url)




# 1) Connect to the database here using the SQLAlchemy's create_engine function
engine = create_engine(url)


# 2) Execute the SQL sentences to create your tables using the SQLAlchemy's execute function
create_table_sql = """
CREATE TABLE table (
    id SERIAL PRIMARY KEY,
    username VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL
)
"""


# 3) Execute the SQL sentences to insert your data using the SQLAlchemy's execute function
with engine.connect() as connection:
    connection.execute(create_table_sql)



# 4) Use pandas to print one of the tables as dataframes using read_sql function
df = pd.DataFrame 
```
